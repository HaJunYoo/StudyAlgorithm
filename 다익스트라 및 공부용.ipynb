```python
import random

import numpy as np

class Graph:
    
    def minDistance(self, dist, queue):
        
        minimum = float("Inf")
        min_index = -1

        for i in range(len(dist)):
            
            if dist[i] < minimum and i in queue:
                minimum = dist[i]
                min_index = i
                
        return min_index

    def printPath(self, parent, j):
        
        if parent[j] == -1:
            print(j, end=' ', flush=True)
            return
        self.printPath(parent, parent[j]) # parent에서 그 parent로 꼬리 물어가는 재귀함수 
        print(j, end=' ', flush=True) 
        # 만약 j=2이라면 2를 프린트하고 2의 부모 노드 번호인 parent[2]를 인쇄하고 그 부모 번호를 또 인쇄하고 종료 조건까지 계속.....
        

    def printSolution(self, dist, parent, minVal, pos):
        
        src = 0
        print("\n최소 거리 노드까지의 경로 : ", end="")
        self.printPath(parent, pos)
        print("")
        
        print("\n전체 경로 살펴보기\n<Vertex>\t\t<Distance from Source>","\t"*4,"<Path>")
        
        for i in range(1, len(dist)):
            #print("\(src} --> {i} \t\t{dist} \t\t\t\t\t".format(src, i, dist[i]), end='\t', flush=True)
            #print('%d %d')
            print("\n%d --> %d \t\t%d \t\t\t\t\t" % (src, i, dist[i]), end='\t', flush=True)
            # 개행을 하지 않고 tab을 한다
            self.printPath(parent, i) # parent 리스트와 i를 받아 printpath 실행
            if (minVal == dist[i]) and (i == pos):
                print("\n","-"*100,"|", end='', flush=True)
            # distance 리스트 안에 i번째 값이 minVal 값과 같고 i가 pos = minindex(최소거리 노드 인덱스가와 같을 때 

    def dijkstra(self, graph, src, seminode):
        
        row = len(graph)
        col = len(graph[0])

        dist = [float("Inf")] * row
        # i까지의 distance를 저장할 리스트는 row의 개수만큼 저장하되 그 원소를 무한대 크기로 한다.

        parent = [-1] * row
        #-1로 초기화시켜 row 개수만큼 parent node list 생성 

        dist[src] = 0

        queue = []
        
        for i in range(row):
            
            queue.append(i)
        # 노드 숫자를 차례대로 큐에 추가 

        while queue:
            
            u = self.minDistance(dist, queue)
            # min index 송출
            
            queue.remove(u)
            #큐에서 min index를 가진 리스트 넘버 제거
            
            for i in range(col):
                
                if graph[u][i] and i in queue:
                    if dist[u] + graph[u][i] < dist[i]:
                        dist[i] = dist[u] + graph[u][i]
                        parent[i] = u
                        
        # graph[u][i]는 u와 i 사이의 노드 엣지라고 보면 된다. dist[u]는 u까지 오는데 쌓인 가중치이다. dist[i]는 i까지 갔을 때 쌓인 가중치. 당연히 i 노드의 부모는 u이다.

        target = [dist[i] for i in seminode]
        # seminode i 값에 해당하는 dist[i] 값들로 이루어져 있는 target 리스트 생성 (리스트 컴프리헨션)
        
        set = {}
        
        for i in range(len(seminode)):
            set.setdefault(seminode[i],target[i])
            
        print("Each pharmacy distance : ", set)
        # 세미노드와 타겟 리스트의 값들을 순서대로 받아로 키와 값으로 set 딕셔너리로 생성 
        
        minVal = min(target)
        # seminode 안 최솟값 구함
        minIdx = np.argmin(target)
        # 인덱스 송출 
        
        print('\nshortest distance pharmacy : ', seminode[minIdx])
        self.printSolution(dist, parent, minVal, seminode[minIdx])
        # dist 리스트, parent 리스트, minVal 값, 최소거리 노드 인덱스 받아 printsolution 실행.

        
    def random_adjacency_matrix(self, n):
        matrix = [[random.randint(0, 500) for i in range(n)] for j in range(n)]

        for i in range(n):
            matrix[i][i] = 0

        for i in range(n):
            for j in range(n):
                matrix[j][i] = matrix[i][j]

        return matrix

g = Graph()

graph = g.random_adjacency_matrix(21)
#시작점을 제외한 20개의 노드 엣지 인접행렬 생성

print("\nGraphs:")

for row in graph :
    print(row)

node = [{'node':1, 'mask':'Dental', 'quantity':3},
        {'node':2, 'mask':'KF94', 'quantity':1},
        {'node':3, 'mask':'KF80', 'quantity':2},
        {'node':4, 'mask':'Fabric', 'quantity':2},
        {'node':5, 'mask':'Dental', 'quantity':4},
        {'node':6, 'mask':'KF94', 'quantity':1},
        {'node':7, 'mask':'Denatal', 'quantity':4},
        {'node':8, 'mask':'KF80', 'quantity':4},
        {'node':9, 'mask':'Fabric', 'quantity':4},
        {'node':10, 'mask':'KF80', 'quantity':1},
        {'node':11, 'mask':'KF94', 'quantity':2},
        {'node':12, 'mask':'Dental', 'quantity':1},
        {'node':13, 'mask':'Fabric', 'quantity':1},
        {'node':14, 'mask':'KF94', 'quantity':4},
        {'node':15, 'mask':'KF80', 'quantity':2},
        {'node':16, 'mask':'Dental', 'quantity':3},
        {'node':17, 'mask':'Fabric', 'quantity':2},
        {'node':18, 'mask':'KF94', 'quantity':5},
        {'node':19, 'mask':'KF80', 'quantity':2},
        {'node':20, 'mask':'KF94', 'quantity':1}]

print("\nNodes:")
for i in node :
    print(i)

def pharmacy():
    
    mask = input("\nInput Mask Type (KF94/KF80/Dental/Fabric) :")
    quant = int(input("Input quantity (1 ~ 5) :"))
    
    seminode = []

    for i in range(0, len(node)):
        if node[i]['mask'] == mask and node[i]['quantity'] >= quant:
            Y = node[i]['node']
            seminode.append(Y)
    
    print('\ncondition satisfying pharmacy : ', seminode)

    try :
        g.dijkstra(graph, 0, seminode)


    except Exception as e:
        print(e)

pharmacy()

```

for 반복문을 두 번 사용하니 코드가 좀 길어졌죠? 리스트 표현식을 활용하면 코드 한 줄로 2차원 리스트를 만들 수 있습니다.

>>> a = [[0 for j in range(2)] for i in range(3)]
>>> a
[[0, 0], [0, 0], [0, 0]]
리스트 표현식 안에서 리스트 표현식을 사용했습니다. 먼저 [0 for j in range(2)]로 0을 2번 반복하여 [0, 0]으로 만들고 다시 for i in range(3)으로 [0, 0]을 3번 반복하여 [[0, 0], [0, 0], [0, 0]]으로 만듭니다.

만약 for 반복문을 한 번만 사용하고 싶다면 다음과 같이 식 부분에서 리스트 자체를 곱해주면 됩니다.

>>> a = [[0] * 2 for i in range(3)]
>>> a
[[0, 0], [0, 0], [0, 0]]
즉, 0이 1개 들어있는 리스트 [0]에 2를 곱하면 [0, 0]이 되는데, 이걸 for i in range(3)으로 3번 반복해주면 [[0, 0], [0, 0], [0, 0]]이 됩니다.
